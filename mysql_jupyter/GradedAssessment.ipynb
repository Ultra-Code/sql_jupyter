```python
# TASK 3
from contextlib import closing

with closing(pool.get_connection()) as cnx:
    cursor= cnx.cursor()
    
    print("The name and EmployeeID of the Little Lemon manager")
    cursor.execute("""
    Select Name, EmployeeID From Employees WHERE Role = "Manager";
    
    """)
    
    results = cursor.fetchall()
    print (cursor.column_names)
    
    for result in results:
        print(result)
    
    print("\n")
    
    print("The name and role of the employee who receives the highest salary")
    cursor.execute("""
    Select Name, Role From Employees WHERE Annual_Salary=(SELECT MAX(Annual_Salary) FROM Employees)
    """)
    
    results = cursor.fetchall()
    print (cursor.column_names)
    
    for result in results:
        print(result)
    
    
    print("\n")
    
    print("The number of guests booked between 18:00 and 20:00.")
    cursor.execute("""
    Select COUNT(BookingID) From Bookings WHERE BookingSlot BETWEEN "18:00:00" and "20:00:00"
    """)
    
    results = cursor.fetchall()
    print (cursor.column_names)
    
    for result in results:
        print(result)
    
    print("\n")
    print("The full name and BookingID of all guests waiting to be seated with the receptionist in sorted order with respect to their BookingSlot")
    cursor.execute("""
    Select CONCAT(GuestFirstName, ' ',GuestLastName) AS FullName, BookingID From Bookings Order By BookingSlot DESC
    """)
    
    results = cursor.fetchall()
    print (cursor.column_names)
    
    for result in results:
        print(result)
```

```python
#TASK 4
basic_sales_report_procedure = """
CREATE PROCEDURE BasicSalesReport()
BEGIN
Select SUM(BillAmount) AS "Total Sales" From Orders;

Select AVG(BillAmount) AS "Average Sales" From Orders;

Select MIN(BillAmount) AS "Minimum Bill Paid" From Orders;

Select MAX(BillAmount) AS "Maximum Bill Paid" From Orders;
END
"""

# Execute the query 
cursor.execute("DROP PROCEDURE IF EXISTS BasicSalesReport")
cursor.execute(basic_sales_report_procedure)

# Call the stored procedure with its name
cursor.callproc("BasicSalesReport")

# Retrieve multiple records in "dataset"
for procedure_result in cursor.stored_results():
    # Retrieve column names using list comprehension in a 'for' loop 
    columns = [ column[0] for column in procedure_result.description ]
    results=procedure_result.fetchall()
    
    # Print column names
    print(columns)
    
    # Print data 
    for result in results:
        print(result)

    print("\n")
```
